```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
new_df = pd.read_csv('Housing (1).csv')
new_df.head(25)
```

How does the per capita crime rate (CRIM) vary across different categories of median home values (CAT. MEDV)?

```python
def get_values(target , arr):
  array = []
  l = new_df['CAT. MEDV'].values
  for i in range(len(arr)):
    if(l[i] == target):
      array.append(arr[i])
  return array
```

```python
plt.figure(figsize = (10,8))
plt.plot(get_values(0.0,new_df['CRIM'].values)[:40],label = 'low_pricing')
plt.plot(get_values(1.0,new_df['CRIM'].values)[:40],label = 'high_pricing')
plt.legend()
plt.show()
print(get_values(0.0,new_df['CRIM'].values))
```

Implementing horizontal bar_chart to show the variation of 1st 25 crime rates


```python

new_val = new_df['CRIM'].iloc[ : 506].str.replace(' ','').astype(float)

new_val.iloc[1:25].plot.barh()
new_val.head()
```

Is there a relationship between the proportion of residential land zoned for lots larger than a certain size (ZN) and the categories of median home values (CAT. MEDV)?

```python
plt.figure(figsize = (10,8))
plt.hist(get_values(0.0,new_df['ZN'].values)[:],label = 'low_pricing')
plt.hist(get_values(1.0,new_df['ZN'].values)[:],label = 'high_pricing')
plt.legend()
plt.show()
```

there doesn't seem any relationship between these two

How does the proportion of non-retail business acres per town (INDUS) differ among the different categories of median home values (CAT. MEDV)?

```python
plt.figure(figsize = (10,8))
plt.subplot(1,2,1)
plt.hist(get_values(0.0,new_df['ZN'].values)[:],label = 'low_pricing')
plt.legend()
plt.subplot(1,2,2)
plt.hist(get_values(1.0,new_df['ZN'].values)[:],label = 'high_pricing')
plt.legend()
plt.show()
```

except for a small fraction the indus area resembles in both the cases but that small franction can be used for the commercial benefit of small bussiness at cheap rate

let's look at the how the NOX values varies in both the cases

```python
plt.figure(figsize = (10,8))
plt.boxplot([get_values(0.0,new_df['NOX'].values),get_values(1.0,new_df['NOX'].values)])
plt.legend()
plt.show()
```

the plots easily shows the worth of high pricing areas as they provide a safe side of living with a low value of NOX( nitric oxides concentration )

let's compare rm , dis,LSTAT and RAD features for both the areas

```python
labels = ['RM' , 'DIS' , 'RAD' , 'LSTAT']
arr = [
    np.array(get_values(0.0,new_df['RM'].values)).mean() ,  np.array(get_values(0.0,new_df['DIS'].values)).mean() ,  np.array(get_values(0.0,new_df['RAD'].values)).mean() ,  np.array(get_values(0.0,new_df['LSTAT'].values)).mean()
]
arr1 = [
    np.array(get_values(1.0,new_df['RM'].values)).mean() ,  np.array(get_values(1.0,new_df['DIS'].values)).mean() ,  np.array(get_values(1.0,new_df['RAD'].values)).mean() ,  np.array(get_values(1.0,new_df['LSTAT'].values)).mean()
]
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.pie(arr , labels = labels , shadow = 'True')
plt.title('low_pricing area')
plt.legend()
plt.subplot(1,2,2)
plt.pie(arr1 , labels = labels , shadow = 'True')
plt.title('high_pricing area')
plt.legend()
```

let's compare the tax and b values in the data set using histograms

```python
plt.figure(figsize = (10,8))
plt.subplot(1,2,1)
plt.hist(get_values(0.0,new_df['B'].values),label = 'B', alpha = 0.4)
plt.hist(get_values(0.0,new_df['TAX'].values),label = 'TAX', alpha = 0.4)
plt.title('low_pricing area')
plt.legend()
plt.subplot(1,2,2)
plt.hist(get_values(1.0,new_df['B'].values),label = 'B', alpha = 0.4)
plt.hist(get_values(1.0,new_df['TAX'].values),label = 'TAX', alpha = 0.4)
plt.title('high_pricing area')
plt.legend()
plt.show()
```

while the low pricing areas are easy to afford and low in the crime rates but they also have a great cost of living making the life afterwards difficult to manage

let's see how the age and the median living range of the areas varies with in both the cateagories

```python
plt.figure(figsize = (10,8))
plt.subplot(1,2,1)
plt.plot(get_values(0.0,new_df['AGE'].values[:20]),label = 'AGE', alpha = 0.4)
plt.plot(get_values(0.0,new_df['MEDV'].values[:20]),label = 'MEDV', alpha = 0.4)
plt.title('low_pricing area')
plt.legend()
plt.subplot(1,2,2)
plt.plot(get_values(1.0,new_df['AGE'].values[:20]),label = 'AGE', alpha = 0.4)
plt.plot(get_values(1.0,new_df['MEDV'].values[:20]),label = 'MEDV', alpha = 0.4)
plt.title('high_pricing area')
plt.legend()
plt.show()
```
