```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
new_df = pd.read_csv('Breast_Cancer (1).csv')
new_df.head(25)
```

let's check the radius range of the radius mean for the people having breast_cancer

```python
range_M = []
range_B = []
val = new_df['diagnosis'].values
val2 = new_df['radius_mean'].values
for i in range(len(val)):
  if val[i] == 'M':
    range_M.append(val2[i])
  else:
    range_B.append(val2[i])

plt.boxplot([range_M,range_B])
plt.show()
```

```python
val = new_df.groupby('diagnosis')
val['texture_mean'].plot.hist(bins = 50 , alpha = 0.5)
```

so, thereby we conclude that the tumors masses are lieng very close range for the women diagnosed with breast cancer

now let's see how the texture, perimeter and the area varies with the diagnosis

```python
def get_values(target_val , arr ):
  val1 = new_df['diagnosis'].values
  array = []
  [array.append(arr[i]) for i in range(len(val1)) if val1[i] == target_val]
  array.sort()
  return array[:-20]

```

now let's see how the texture, perimeter and the area varies with the diagnosis

```python
plt.figure(figsize = (15,8))
plt.subplot(1,2,1)
plt.plot(get_values('M',new_df['texture_mean'].values),label = 'texture')
plt.plot(get_values('M',new_df['perimeter_mean'].values),label = 'perimeter')
plt.plot(get_values('M',new_df['area_mean'].values) , label = 'area')
plt.legend(title = 'lines_cataegories')
plt.title('for healthy females')
plt.subplot(1,2,2)
plt.plot(get_values('B',new_df['texture_mean'].values),label = 'texture')
plt.plot(get_values('B',new_df['perimeter_mean'].values),label = 'perimeter')
plt.plot(get_values('B',new_df['area_mean'].values),label = 'area')
plt.title('for females diagnosed with breast cancer')
plt.legend(title = 'lines_cataegories')
plt.suptitle('values fo these attributes for highest count in 20 females for btoh the cataegories')
plt.show()
```

```python
val = new_df.groupby('diagnosis')
val[['texture_mean','area_mean','perimeter_mean']].plot.area()
```

the above graph shows that the 3 features decreases mearly 50% in case of females diagnosed with the breast cancer, Thus affecting their appearence at a very high rate

let's check out how the smoothness is varing with the disease

```python
plt.plot(get_values('M',new_df['smoothness_mean'].values),label = 'healthy')
plt.plot(get_values('B',new_df['smoothness_mean'].values),label = 'diseased')
plt.legend(title = 'types')
plt.title('smoothness')
plt.show()
```

```python
plt.figure(figsize = (10,5))
val = new_df.groupby('diagnosis')
val['smoothness_mean'].plot.hist(bins = 150 , alpha = 0.5)
```

the graph shows that the smoothness mean in the healthy females ranges somewhere between 0.7 to 1.2, whereas in the diseased it starts from 0.5 which is the lowest among top 20 which makes a lot of variation for just a simple cataegory

let's see how it this disease makes it's impact upon other components and how they are varing in both the cases


```python
plt.figure(figsize = (15,8))
plt.subplot(1,2,1)
plt.plot(get_values('M',new_df['compactness_mean'].values),label = 'compactness')
plt.plot(get_values('M',new_df['concavity_mean'].values),label = 'concavity')
plt.plot(get_values('M',new_df['concave points_mean'].values) , label = 'cancave point')
plt.plot(get_values('M',new_df['symmetry_mean'].values) , label = 'symmetry point')
plt.plot(get_values('M',new_df['fractal_dimension_mean'].values) , label = 'fractal_dimenson point')
plt.legend(title = 'lines_cataegories')
plt.title('for healthy females')
plt.subplot(1,2,2)
plt.plot(get_values('B',new_df['compactness_mean'].values),label = 'compactness')
plt.plot(get_values('B',new_df['concavity_mean'].values),label = 'concavity')
plt.plot(get_values('B',new_df['concave points_mean'].values) , label = 'cancave point')
plt.plot(get_values('B',new_df['symmetry_mean'].values) , label = 'symmetry point')
plt.plot(get_values('B',new_df['fractal_dimension_mean'].values) , label = 'fractal_dimenson point')
plt.legend(title = 'lines_cataegories')
plt.title('for cancer diagnosed females')
plt.suptitle('jboiehreir')
plt.show()
```

```python
val = new_df.groupby('diagnosis')
val['compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean'].plot.area()
```

the above graphs shows that all other features except for the fractral dimensions is effective swerely
