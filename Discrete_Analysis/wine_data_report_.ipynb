```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
new_df = pd.read_csv('wine_data (1).csv')
new_df.head(25)
```

note* I tried to show plotting graphs using both pandas and matplotlib for better understanding of functionality of both the libraries

let's see how the alcoholic contents vary according to the wine types

```python
def get_values(wine_type, arr):
  array = []
  value = new_df['Type'].values
  for i in range(len(arr)):
    if value[i] == wine_type:
      array.append(arr[i])
  return array
```

```python
plt.hist(get_values(1,new_df['Alcohol'].values),label = 'type-1',alpha = 0.7,width = 0.2)
plt.hist(get_values(2,new_df['Alcohol'].values),label = 'type-2',alpha = 0.7,width = 0.2)
plt.hist(get_values(3,new_df['Alcohol'].values),label = 'type-3',alpha = 0.7,width = 0.2)
plt.title('variation in alcoholic content of different types of wines')
plt.legend()
plt.show()
```

```python
new_val = new_df.groupby('Type')
new_val['Alcohol'].plot.hist(bins = 18, alpha = 0.5)
```

this graph shows that the alcoholic content in the type is minimal among the 3 followed by type 3 and the highest one in the type 1

let's see how the malic acid and the ash {this is basically the measure between the organic and the inorganic conetnt in the wine}varies in the type of wines

```python
plt.figure(figsize =  (15,8))

plt.subplot(1,3,1)
new_arr = [get_values(1,new_df['Malic']), get_values(1,new_df['Ash'])]
plt.boxplot(new_arr)
plt.title('comperison for type-1')
plt.subplot(1,3,2)
new_arr = [get_values(2,new_df['Malic']), get_values(2,new_df['Ash'])]
plt.boxplot(new_arr)
plt.title('comperison for type-2')
plt.subplot(1,3,3)
new_arr = [get_values(3,new_df['Malic']), get_values(3,new_df['Ash'])]
plt.boxplot(new_arr)
plt.title('comperison for type-3')
plt.suptitle('comperison between the organic and inorganic values of wines')
plt.show()
```

```python
new_val = new_df.groupby('Type')
new_val['Malic' , 'Ash'].plot.box()
```

from the interquartile range of the boxplots we can see the most appropriate values of organic contents is present in the type-3 wines and thus, is safe for consuption among these 3 types

now here is the check for the alcalinity and the magnesium contents in them

```python
plt.figure(figsize = (15,8))
plt.subplot(1,3,1)
plt.plot(get_values(1,new_df['Alcalinity']),label = 'Alcalinity')
plt.plot(get_values(1,new_df['Magnesium']),label = 'Magnesium')
plt.title('contents for type-1')
plt.legend()
plt.figure(figsize = (15,8))
plt.subplot(1,3,3)
plt.plot(get_values(2,new_df['Alcalinity']),label = 'Alcalinity')
plt.plot(get_values(2,new_df['Magnesium']),label = 'Magnesium')
plt.title('contents for type-2')
plt.legend()
plt.figure(figsize = (15,8))
plt.subplot(1,3,3)
plt.plot(get_values(3,new_df['Alcalinity']),label = 'Alcalinity')
plt.plot(get_values(3,new_df['Magnesium']),label = 'Magnesium')
plt.title('contents for type-3')
plt.legend()
plt.suptitle('Magnesium and alcalinity checks')
```

```python
new_val = new_df.groupby('Type')
new_val['Alcalinity','Magnesium'].plot.area()
```

now here's the check for the comosition of organic compounds in the following types

```python
plt.figure(figsize = (15,8))
plt.subplot(1,3,1)
arr = [np.mean(get_values(1,new_df['Phenols'].values)),np.mean(get_values(1,new_df['Flavanoids'].values)),np.mean(get_values(1,new_df['Nonflavanoids'])),np.mean(get_values(1,new_df['Proanthocyanins'].values))]
plt.pie(arr,labels = ['Phenols','Flavanoids','Nonflavanoids','Proanthocyanins'],shadow = 'True')
plt.title('type-1')
plt.subplot(1,3,2)
arr2 = [np.mean(get_values(2,new_df['Phenols'].values)),np.mean(get_values(2,new_df['Flavanoids'].values)),np.mean(get_values(2,new_df['Nonflavanoids'])),np.mean(get_values(2,new_df['Proanthocyanins'].values))]
plt.pie(arr2,labels = ['Phenols','Flavanoids','Nonflavanoids','Proanthocyanins'],shadow = 'True')
plt.title('type-2')
plt.subplot(1,3,3)
arr3 = [np.mean(get_values(3,new_df['Phenols'].values)),np.mean(get_values(3,new_df['Flavanoids'].values)),np.mean(get_values(3,new_df['Nonflavanoids'])),np.mean(get_values(3,new_df['Proanthocyanins'].values))]
plt.pie(arr3,labels = ['Phenols','Flavanoids','Nonflavanoids','Proanthocyanins'],shadow = 'True')
plt.title('type-3')
plt.suptitle('organic composition')
plt.show()
```

now, let's checkout the flavouring and the freshiening agents comorsition

```python
plt.figure(figsize = (15,8))
plt.subplot(1,3,1)
arr = [np.mean(get_values(1,new_df['Color'].values)),np.mean(get_values(1,new_df['Hue'].values)),np.mean(get_values(1,new_df['Dilution']))]
plt.pie(arr,labels = ['Color', 'Hue' , 'Dilution' ],shadow = 'True')
plt.title('type-1')
plt.subplot(1,3,2)
arr2 = [np.mean(get_values(2,new_df['Color'].values)),np.mean(get_values(2,new_df['Hue'].values)),np.mean(get_values(2,new_df['Dilution']))]
plt.pie(arr2,labels = ['Color', 'Hue' , 'Dilution' ],shadow = 'True')
plt.title('type-2')
plt.subplot(1,3,3)
arr3 = [np.mean(get_values(3,new_df['Color'].values)),np.mean(get_values(3,new_df['Hue'].values)),np.mean(get_values(3,new_df['Dilution']))]
plt.pie(arr3,labels = ['Color', 'Hue' , 'Dilution' ],shadow = 'True')
plt.title('type-3')
plt.suptitle('organic composition')
plt.show()
```

```python
new_val = new_df.groupby('Type').agg(np.mean)
print(new_val)
new_val[['Color','Hue' , 'Dilution']].plot.pie(subplots = True , figsize = (15,8),fontsize = (25))
```

now, let's compare the aroma and the flavours of these 3 types of wines

```python
plt.hist(np.array(get_values(1,new_df['Proline'].values)),label = 'type-1' , alpha = 0.5)
plt.hist(np.array(get_values(2,new_df['Proline'].values)),label = 'type-2' , alpha = 0.5)
plt.hist(np.array(get_values(3,new_df['Proline'].values)),label = 'type-3' , alpha = 0.5)
plt.title('proline status in different types of wines :')
plt.legend(title = 'type_recognistion')
plt.show()
```
